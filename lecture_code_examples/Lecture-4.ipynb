```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def figure(figsize=(6, 5), xlim=None, ylim=None, title=None):
    plt.figure(figsize=figsize)
    plt.grid(linestyle='--')
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
    if title is not None:
        plt.title(title, fontsize=16)
```

## Open Methods (Fixed Point Iteration)

```python
# f(x) = x**3 - a - 0
# g(x) = x + C*(x**3 - a)
a = 2
n = 10
g = 1.
C = -0.21

sq = np.zeros(n)
sq[0] = g
for i in range(1, n):
    sq[i] = C * (sq[i - 1] ** 3 - a) + sq[i - 1]
figure(xlim=(0, 10), ylim=(-0.4, 1.4))
plt.plot([0, n], [a**(1./3.), a**(1./3.)], 'b', label='Exact')
plt.plot(np.arange(1, n + 1), sq, 'r', label='Iteration')
plt.plot(np.arange(1, n + 1), (sq - a ** (1. / 3.)) / (a ** (1. / 3.)), 'lime', label='Error')
plt.xticks(np.arange(11))
plt.title(f'a = {a}, C = {C}', fontsize=16)
plt.legend();
```

## Newton-Raphson Method: Example

```python
a = 26
n = 10
g = 1
sq = np.zeros(n)
sq[0] = g

for i in range(1, n):
    sq[i] = 0.5 * (sq[i - 1] + a / sq[i - 1])

figure(xlim = (0, 10), ylim=(-5, 30))
sqrta = np.sqrt(a)
plt.plot([0, n], [sqrta, sqrta], 'b')
x = np.arange(1, n + 1)
plt.plot(x, sq, 'r')
plt.plot(x, a/sq, 'r-.')
plt.plot(x, (sq - sqrta) / sqrta, 'lime')
plt.legend(['Exact', 'sq', 'a/sq', 'Error']);
```

## Newton-Raphson Example: Its use for divisions

```python
a = 10
n = 10
g = 0.19
sq = np.zeros(n)
sq[0] = g
for i in range(1, n):
    sq[i] = sq[i - 1] - sq[i - 1] * (a * sq[i - 1] - 1)
    
figure(xlim=[0, 10], ylim=[-1, 1], title=f'x = 1 / {a}'); 

plt.plot([0, n], [1/a, 1/a], 'b')
x = np.arange(1, n + 1)
plt.plot(x, sq, 'r')
plt.plot(x, (sq - 1 / a) * a, 'lime')
plt.legend(['Exact', 'Iteration', 'Rel Error'])
```
